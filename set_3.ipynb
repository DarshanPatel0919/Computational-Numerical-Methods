```python
import numpy as np
import pandas as pd
from scipy.misc import derivative as drv
import matplotlib.pyplot as plt

st=-2.5
en=2.5
dt=1e-2
eps=1e-5
dx=1e-9
n =int((en-st)/dt) + 1
x=np.linspace(st,en,n)
def f1(x):
    return x**6 - x - 1
def f2(x):
    return x**3 - x**2 - x - 1
def f3(x):
    return x - 1 - (0.3)*np.cos(x)
def f4(x):
    return np.cos(x)-np.sin(x)-0.5
def f5(x):
    return x - np.exp(-x)
def f6(x):
    return np.sin(x) - np.exp(-x)
def f7(x):
    return x**3 - 2*x - 2
def f8(x):
    return x**4 - x - 1
def f9(x):
    return np.exp(x) - x - 2
def f10(x):
    return 1 - x + np.sin(x)
def f11(x):
    return np.tan(x)-x
 
def f(x,i):
    if i==1:
        return f1(x)
    elif i==2:
        return f2(x)
    elif i==3:
        return f3(x)
    elif i==4:
        return f4(x)
    elif i==5:
        return f5(x)
    elif i==6:
        return f6(x)
    elif i==7:
        return f7(x)
    elif i==8:
        return f8(x)
    elif i==9:
        return f9(x)
    elif i==10:
        return f10(x)
    elif i==11:
        return f11(x)
    
for i in range(11):
    plt.figure(i)
    plt.plot(x,f(x,i+1))
    
    
```
