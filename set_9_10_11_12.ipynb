```
import numpy as np
import pandas as pd
import pylab as plt
import six

precision=6
eps=10**(-precision)
c=0
prec='0.'+str(precision)+'f'
matrix=[]
def jacobi(a,b,x0):
  global c
  c+=1
  x1=[]
  l=[c]
  for i in range(len(b)):
    x1.append( (b[i,0] - np.sum(np.matmul(a[i,:],x0)) + a[i][i]*x0[i,0])/a[i][i])
    l.append(format(x1[i],prec))
  matrix.append(l)
  x1=np.transpose(np.array([x1]))
  if(np.sum(np.absolute(x1-x0))<eps):
    return x1
  return jacobi(a,b,x1)


def gauss_seidel(a,b,x0):
  global c
  c+=1
  x1=[]
  temp=np.copy(x0)
  l=[c]
  for i in range(len(b)):
    x0[i,0]=float((b[i,0] - np.sum(np.matmul(a[i,:],x0)) + a[i][i]*x0[i,0])/a[i][i])
    l.append(format(x0[i,0],prec))
  matrix.append(l)
  if(np.sum(np.absolute(temp-x0))<eps):
    return x0
  return gauss_seidel(a,b,np.copy(x0))

def render_mpl_table(data, col_width=3.0, row_height=0.625, font_size=14,
                     header_color='#40466e', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in six.iteritems(mpl_table._cells):
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax

b=np.transpose(np.array([[10,19,0]]))
a = np.array([[9,1,1],[2,10,3],[3,4,11]])
  
  
  
x0=np.transpose(np.array([[0.0,0.0,0.0]]))

x = np.matmul(np.linalg.inv(a),b)
c=0
names =['n']
for i in range(len(b)):
  names.append('x('+str(i)+')')
print(jacobi(a,b,x0),c)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)
c=0
matrix=[]
print(gauss_seidel(a,b,x0),c)
print(x)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)

x0=np.transpose(np.array([[2000.0,2000.0,2000.0]]))

x = np.matmul(np.linalg.inv(a),b)
c=0
names =['n']
for i in range(len(b)):
  names.append('x('+str(i)+')')
matrix=[]
print(jacobi(a,b,x0),c)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)
c=0
matrix=[]
print(gauss_seidel(a,b,x0),c)
print(x)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)


x0=np.transpose(np.array([[-20.0,-1000.0,-530.0]]))

x = np.matmul(np.linalg.inv(a),b)
c=0
names =['n']
for i in range(len(b)):
  names.append('x('+str(i)+')')
matrix=[]
print(jacobi(a,b,x0),c)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)
c=0
matrix=[]
print(gauss_seidel(a,b,x0),c)
print(x)
df = pd.DataFrame(np.asarray(matrix),columns=names)
p1=render_mpl_table(df, header_columns=0, col_width=2.0)

```

```
import numpy as np
import pandas as pd
import pylab as plt
import six

dx=1e-9
eps=1e-4
def f0(x):
  return x[0]**2 + 4*x[1]**2 - 9
def f1(x):
  return 18*x[1] - 14*x[0]**2 + 45
def f(x,i):
  if(i):return f1(x)
  else:return f0(x)
def ff(x,i,j):
  t = np.array([ x[k]+dx if (k==j) else  x[k]  for k in range(len(x))])
  return (f(t,i)-f(x,i))/dx

def gen_newton(x0):
    F = np.array([[ff(x0,i,j) for j in range(len(x0))]for i in range(len(x0))])
    #print(F.shape)
    Fi = np.transpose(np.array([f(x0,i) for i in range(len(x0))]))
    x1 = x0 - np.matmul(np.linalg.pinv(F),Fi)
    if(np.sum(np.absolute(x1-x0))<eps):return x0
    return gen_newton(np.copy(x1))

  
  
x0=np.transpose(np.array([0.50,-0.50]))
roots = gen_newton(x0)
print(roots)
x0=np.transpose(np.array([-0.50,-0.50]))
roots = gen_newton(x0)
print(roots)
x0=np.transpose(np.array([-0.50,0.50]))
roots = gen_newton(x0)
print(roots)
x0=np.transpose(np.array([0.50,0.50]))
roots = gen_newton(x0)
print(roots)
```

```
import numpy as np
import pandas as pd
import pylab as plt
import six

def f1(x,y):
  return -y
def f2(x,y):
  return (y+x**2-2)/(1+x)

def euler1(h,y0,l):
  y=[y0]
  for n in range(l-1):
    y.append(y[len(y)-1]+h*f1(n*h,y[len(y)-1]))
  return y

def euler2(h,y0,l):
  y=[y0]
  for n in range(l-1):
    y.append(y[len(y)-1]+h*f2(n*h,y[len(y)-1]))
  return y
   
st,en=0,6
h=[0.2,0.1,0.05]
error=[]
l = int((en-st)/h[0])+1
x=np.linspace(st,en,l)
y = np.exp(-x)
plt.plot(x,y,'r--',label='Analytical')
for i in range(3):
  l = int((en-st)/h[i])+1
  x=np.linspace(st,en,l)
  y = euler1(h[i],1,l)
  ya = np.exp(-x)
  y,ya=np.array(y),np.array(ya)
  error.append(np.absolute(y-ya))
  plt.plot(x,y,label='h = '+str(h[i]))
  

plt.legend(loc='best')
plt.show()

for i in range(3):
  l = int((en-st)/h[i])+1
  x=np.linspace(st,en,l)
  plt.plot(x,error[i],label='h = '+str(h[i]))
plt.legend(loc='best')  
plt.show()




### Q2 ###

error1=[]
x=np.linspace(st,en,l)
y = (x**2 + 2*x + 2 - 2*(x+1)*np.log(x+1))
plt.plot(x,y,'r--',label='Analytical')
for i in range(3):
  l = int((en-st)/h[i])+1
  x=np.linspace(st,en,l)
  y = euler2(h[i],2,l)
  ya =(x**2+2*x+2-2*(x+1)*np.log(x+1))
  y,ya=np.array(y),np.array(ya)
  error1.append(np.absolute(y-ya))
  plt.plot(x,y,label='h = '+str(h[i]))
  

plt.legend(loc='best')
plt.show()

for i in range(3):
  l = int((en-st)/h[i])+1
  x1=np.linspace(st,en,l)
  plt.plot(x1,error1[i],label='h = '+str(h[i]))
plt.legend(loc='best')  
plt.show()
```
